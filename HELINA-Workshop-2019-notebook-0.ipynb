## The Cholera outbreak in London in 1854 forever changed our perspectives on 

## the importance of Public Health, Epidemiology, Science, and Water / Waste water

## management within growing urban cities and rural communities .

## The Miasma theory of 1854 held that "bad air" and "stench" were to blame for the deaths.

## There was a lot of misinformation and a lack of scientific method within 1854 London.

## A physician, Dr. John Snow, mapped the deaths and the proximity to water pumps.

## His systematic approach - involved what became known as "shoe-leather epidemiology".

## He went door-to-door to visit homes and map the deaths and interview families.

## He mapped the water sources and pumps used by each household.

## He retraced the steps of the mothers that often gathered the water and washed diapers.

## There were areas where diapers were washed that were in close proximity to the well for the Broad Street Pump.

## The London households disposed of waste in the cellars and pits outside the homes.

## Dr. Snow began to suspect the waste was contaminating the drinking water sources.

## Many deaths were proximate to the Broad Street pump.   They removed the pump handle.

## The outbreak stopped and the modern science of Epidemiology was formed.
