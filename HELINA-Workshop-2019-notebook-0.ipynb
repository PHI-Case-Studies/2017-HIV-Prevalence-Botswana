The Cholera outbreak in London in 1854 forever changed our perspectives on The Cholera outbreak in London in 1854 forever changed our perspectives on the importance of public health, epidemiology, science, and water / waste water management within growing urban cities and rural communities.

The Miasma theory of 1854 held that "bad air" and "stench" were to blame for the deaths. There was a lot of misinformation and a lack of scientific method within 1854 London. A physician, Dr. John Snow, mapped the deaths and the proximity to water pumps. His systematic approach - involved what became known as "shoe-leather epidemiology". He went door-to-door to map the deaths and interview families. He mapped the water sources, pumps used by each household and retraced the steps of the mothers that often gathered the water and washed diapers. There were areas where diapers were washed that were in close proximity to the well for the Broad Street Pump. In addition, London households disposed of waste in their cellars and pits outside the homes, which were also close to the well.

Dr. Snow began to suspect the waste was contaminating the drinking water sources.
Many deaths were proximate to the Broad Street pump. They removed the pump handle.
The outbreak stopped and the modern science of epidemiology was formed.
